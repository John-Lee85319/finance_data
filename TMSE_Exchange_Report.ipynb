```python
# 套件載入
import requests
from io import StringIO
import pandas as pd
import numpy as np
import time
```

```python
# 爬取證交所
# https://medium.com/%E5%8F%B0%E8%82%A1etf%E8%B3%87%E6%96%99%E7%A7%91%E5%AD%B8-%E7%A8%8B%E5%BC%8F%E9%A1%9E/%E7%A8%8B%E5%BC%8F%E8%AA%9E%E8%A8%80-%E8%82%A1%E7%A5%A8%E7%88%AC%E8%9F%B2-b502af0e3de7

# 時間序列日期打印
# https://kknews.cc/zh-tw/finance/yz354vg.html

# 資料合併
# https://ithelp.ithome.com.tw/articles/10200433
```

```python
# 輸入要抓的時間日期
day=pd.period_range('2010-07-01','2020-07-01',freq='d')
dayline = []
remap ={ord('-'): None}
for z in range(len(day)):
    container = str(day[z])
    container = container.translate(remap)
    dayline.append(container)
```

```python
# 創立一個空的 dataframe
df_empty = pd.DataFrame(columns=['Unnamed: 0', '證券代號', '證券名稱', '成交股數', '成交筆數', '成交金額', '開盤價', '最高價',
       '最低價', '收盤價', '漲跌(+/-)', '漲跌價差', '最後揭示買價', '最後揭示買量', '最後揭示賣價', '最後揭示賣量',
       '本益比'])
```

```python
# 爬取期交所網站，爬的過程中會告知那一天沒有交易
for u in range(len(dayline)):
    r = requests.post('https://www.twse.com.tw/exchangeReport/MI_INDEX?response=csv&date=' + dayline[u] + '&type=ALL')
    if r.text == '':
        print("這天期交所沒開盤:" + dayline[u])
    else:
        space_dict={ord(c): None for c in ' '}
        split_r = r.text.split('\n')
        function = [i.translate(space_dict)
                   for i in split_r
                   if len(i.split('",'))==17 and i[0] != '=' ]
        fuction_end ="\n".join(function)
        function_answer = StringIO(fuction_end)
        df = pd.read_csv(function_answer,header=0)
        df = df.drop(labels=['Unnamed: 16'],axis='columns')
        df.to_csv('C:\\Users\\Robin\\Desktop\\data\\'+dayline[u]+'.csv')
        time.sleep(2)
```

```python
# 開啟檔案並且將檔案合併
day_close = []
for v in range(len(dayline)):
    try:
        df_csv =pd.read_csv('C:\\Users\\Robin\\Desktop\\data\\'+ dayline[v] +'.csv')
        df_empty =pd.concat([df_empty,df_csv],axis=0)
    except FileNotFoundError:
        print(dayline[v])
        day_close.append(dayline[v])
df_empty = df_empty.drop(labels=['Unnamed: 0'],axis='columns')
```
